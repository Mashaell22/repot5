```python
data = {
    'Gender': ['Male', 'Male', 'Female', 'Female', 'Male', 'Female', 'Female', 'Female', 'Male', 'Male'],
    'Height': [174, 189, 185, 195, 149, 150, 184, 141, 150, 173],
    'Weight': [96, 87, 110, 104, 61, 153, 121, 136, 95, 131],
    'Index': [4, 2, 4, 3, 3, 5, 4, 5, 5, 5]
}
df = pd.DataFrame(data)


print(df.head())
```

```python
def categorize_height(height):
    if height < 150:
        return 'Short'
    elif 150 <= height < 170:
        return 'Average'
    elif 170 <= height < 190:
        return 'Tall'
    else:
        return 'Very Tall'


df['Height_Category'] = df['Height'].apply(categorize_height)


print(df.head())
```

```python
def categorize_bmi(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 25:
        return 'Normal weight'
    elif 25 <= bmi < 30:
        return 'Overweight'
    elif 30 <= bmi < 35:
        return 'Obese Class I'
    elif 35 <= bmi < 40:
        return 'Obese Class II'
    else:
        return 'Obese Class III'


df['BMI_Category'] = df['BMI'].apply(categorize_bmi)


print(df.head())
```

This data set have 4 Coulmns Gender , height , weight and index.
if we want to create a new feature , we should do some mathmatical opreation in the data set to create a new feature (new column)


### **what about bmi? (عامود لحساب كتلة الجسم)**

### **bmi = weight ÷ Height^2 (in meter)**
the height in this data set not in meter
so we need to convert it from CM to M
to do this we need to create new coulmn :

### **height / 100 **


```python
df['Height_M'] = df['Height'] / 100
```

### **Let's see what we got!**

```python
df.head()
```

Now we can create new feature from this dataset.
as we mentioned before we want to caluclate bmi

### **bmi = weight ÷ Height^2 (in meter)**


```python
df['BMI'] = df['Weight'] / (df['Height_M'] ** 2)
```
